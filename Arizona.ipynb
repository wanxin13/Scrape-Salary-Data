```python
from bs4 import BeautifulSoup
```

```python
from urllib.request import urlopen
```

```python
last_quote_page = 'https://www.openthebooks.com/search/?PensionCode=801&perpage=500&F_fiscalyear=2017&F_employernm=&F_employeenm=&pg=404'
```

```python
periods = []
for i in range(0,342):      # Number of pages plus one 
    url = "https://www.openthebooks.com/search/?PensionCode=801&F_fiscalyear=2018&F_employernm=&F_employeenm=&perpage=500&pg={}" .format(i)
    page = urlopen(url)
    soup = BeautifulSoup(page,'html.parser')
    a = soup.find_all('td', class_='tdsite')
    period_tags = a
    periods.append([pt.get_text() for pt in period_tags])
```

```python
p = []
for i in range(1,342):
    p = p + periods[i]

e = len(p)
CO = pd.DataFrame({
"Year": p[1:e-4:6],
"Name": p[4:e-1:6],
"Dept": p[2:e-3:6],
"Title": p[3:e-2:6],
"Salary": p[5:e:6]
})
CO
```
