```python
periods = []
for i in range(0,42):      # Number of pages plus one 
    url = "https://www.openthebooks.com/search/?PensionCode=4601&F_fiscalyear=2018&F_employernm=&F_employeenm=&perpage=500&pg={}" .format(i)
    page = urlopen(url)
    soup = BeautifulSoup(page,'html.parser')
    a = soup.find_all('td', class_='tdsite')
    period_tags = a
    periods.append([pt.get_text() for pt in period_tags])
    print(i)
```
