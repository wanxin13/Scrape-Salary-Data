```python
periods = []
periods1 = []
for i in range(1,2395):      # Number of pages plus one 
    url = "http://db.desmoinesregister.com/state-salaries-for-iowa/page={}&ordercol=col2&orderdir=asc&searchterms%5Bcol3%5D=&searchterms%5Bcol2%5D=&searchterms%5Bcol6%5D=&searchterms%5Bcol11%5D=2014" .format(i)
    page = urlopen(url)
    soup = BeautifulSoup(page,'html.parser')
    a = soup.find_all('tr', class_='datarow-odd')
    period_tags = a
    periods.append([pt.get_text() for pt in period_tags])
    b = soup.find_all('tr', class_='datarow-even')
    period_tags = b
    periods1.append([pt.get_text() for pt in period_tags])
```
